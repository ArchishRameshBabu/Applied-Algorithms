```python
import random
```

```python
class treeNode:
    def __init__(self,key,ht):
        self.parent=None
        self.lc=None
        self.rc=None
        self.data=key
        self.height=ht
        self.priority=random.random()
```

```python
class Treap:


    def insert(self,root,val):
        if (root is None):
            print("No need")
        
        
        if root.data >= val:
            
            if (root.lc == None):
                
                nlc=treeNode(val,0)
                nlc.parent = root
                root.lc = nlc
                self.updateHeight(root)
                
                while(root.parent is not None):
                    root=root.parent
                
                
                while(nlc!=root and nlc.priority < nlc.parent.priority):
                    #print("Inside the while loop")
                    #print(nlc.parent)
                    root=(self.rotate_right(nlc,root))
                    #print(nlc.parent)
                
                #print("the new root",root.data)
                #return(root)
            
            else:
                self.insert(root.lc, val)
        
        else:
            if (root.rc == None):
                nlc=treeNode(val,0)
                nlc.parent = root
                root.rc = nlc
                self.updateHeight(root)
                
                
                while(root.parent is not None):
                    root=root.parent
                    
                while(nlc is not root and nlc.priority < nlc.parent.priority):
                    #print("Inside the while loop")
                    #print(nlc.parent)
                    root=(self.rotate_left(nlc,root))
                    #print(nlc.parent)
                
                
                #print("the new root",root.data)
                #return(root)
            else:
                self.insert(root.rc, val)
        
    def inorder(self,root):
        if root:
            self.inorder(root.lc)
            print("The data values is=",root.data,"The height is=",root.height,"The priority is=",root.priority)
            self.inorder(root.rc)

            
    def search(self,root,key):
        if root is None:
            return None
        if root.data==key:
            return root
        elif root.data<key:
            return(self.search(root.rc,key))
        else:
            return(self.search(root.lc,key))
        
        
        
    def updateHeight(self,root):
        if (root == None):
            return "No Height"

        oldHeight = root.height
        if(root.lc is None):
            lc_ht=0
        else:
            lc_ht=root.lc.height
        
        if(root.rc is None):
            rc_ht=0
        else:
            rc_ht=root.rc.height
        
        newHeight = 1 + max(lc_ht,rc_ht)

        if (oldHeight == newHeight):
            return
        else:
            root.height = newHeight
            self.updateHeight(root.parent)
            
    def siftUp(self,x,root):
        if (x.parent == None):
            return root
        while(x.parent.rc is not None and x==x.parent.rc and x.priority < x.parent.priority):
            root=(self.rotate_left(x,root))
        while(x.parent.lc is not None and x==x.parent.lc and x.priority < x.parent.priority):
            root=(self.rotate_right(x,root))
        return root
    
    def rotate(self,x,root):
        if (x == root):
            return root
        elif (x==x.parent.rc):
            return(self.rotate_left(x,root))
        else:
            return(self.rotate_right(x,root))
        
        
            
    def rotate_right(self,x,root):
        y = x.parent
        z = x.rc
        yParent = y.parent

        if (yParent is None):
            root=x
        elif(yParent.lc == y):
                yParent.lc = x
        else:
            yParent.rc = x

        x.parent = yParent
        x.rc = y

        y.parent = x
        y.lc = z

        if (z is not None):
            z.parent = y

        self.updateHeight(x)
        self.updateHeight(y)
        return(root)
        
    def rotate_left(self,x,root):
        y = x.parent
        z = x.lc
        yParent = y.parent

        if (yParent is None):
            root=x
        elif(yParent.rc == y):
                yParent.rc = x
        else:
            yParent.lc = x

        x.parent = yParent
        x.lc = y

        y.parent = x
        y.rc = z

        if (z is not None):
            z.parent = y

        self.updateHeight(x)
        self.updateHeight(y)
        return(root)

        
```

```python
for i in range(5):
    treap=Treap()
    root = treeNode(0,0)

    for element in range(1,100000):
        treap.insert(root, element)
        while(root.parent is not None):
            root=root.parent

    while(root.parent is not None):
        root=root.parent
    
    print("Iteration",i+1)    
    print("Root data, priority, height: ", root.data,root.priority,root.height)
    print("Root's left child data, priority, height: ",root.lc.data,root.lc.priority,root.lc.height)
    print("Root's right child data, priority, height: ",root.rc.data,root.rc.priority,root.rc.height)
    
```
