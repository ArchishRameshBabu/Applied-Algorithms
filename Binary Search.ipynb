```python
class treeNode:
    def __init__(self,key):
        self.parent=None
        self.lc=None
        self.rc=None
        self.data=key   
```

```python
class BST:

    def insert(self,root,val):
        if (root is None):
            print("No need")
        
        if root.data >= val:
            
            if (root.lc == None):
                
                nlc=treeNode(val)
                #nlc.value = val
                nlc.parent = root
                
                root.lc = nlc
            
            else:
                self.insert(root.lc, val)        
        else:
            if (root.rc == None):
                
                nlc=treeNode(val)
                #nlc.value = val
                nlc.parent = root
                root.rc = nlc
            
            else:
                self.insert(root.rc, val)
        
    def inorder(self,root):
        if root: 
            self.inorder(root.lc)
            print(root.data)
            self.inorder(root.rc)
            
    def search(self,root,key):
        if root is None:
            return False
        if root.data==key:
            return True
        elif root.data<=key:
            return(self.search(root.rc,key))
        else:
            return(self.search(root.lc,key))
```
